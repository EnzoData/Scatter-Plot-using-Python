```python
%matplotlib inline
import pandas as pd

# Suppress warnings
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/datasets/mtcars.csv', index_col=0)
```

```python
df.head()
```

```python
df.plot.scatter(x='wt', y='mpg')
```

```python
import seaborn as sns

# "Styling"
sns.set_style('white')
sns.set_style('ticks')

# Using regplot to create the scatter plot with 95% confidence interval and a regression line
sns.regplot(x='wt', y='mpg', data=df)
```

```python
sns.regplot(x='wt', y='mpg', color='g', data=df)
```

```python
sns.lmplot(x='wt', y='mpg', data=df)
```

```python
sns.lmplot(x='wt', y='mpg', hue='vs', data=df)
```

```python
sns.lmplot(x='wt', y='mpg', hue='vs', palette="Set1", data=df)
```

```python
sns.lmplot(x='wt', y='mpg', hue='vs',palette="Set1", markers='+', data=df)
```

```python
sns.lmplot(x='wt', y='mpg', hue='vs',palette="Set1", markers=['+', 'o'], data=df)
```
